```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('tabla_secuestros')
```

```python
DF=pd.read_json("https://www.datos.gov.co/resource/d7zw-hpf4.json?$limit=100000")
DF
```

```python
def ejecutar_sql(query, conexion=conn):
  """
  Toma una consulta SQL, la ejecuta y devuelve el resultado
  en una tabla de pandas para una visualización clara.
  """
  df = pd.read_sql_query(query, conexion)
  # Usamos display() porque en Colab presenta las tablas de forma más elegante.
  display(df)
```

```python
DF.to_sql("secuestros", conn)
```

```python
ejecutar_sql("SELECT tipo_delito, SUM(cantidad) as total FROM secuestros GROUP BY tipo_delito")
```

```python
ejecutar_sql("SELECT municipio, SUM(cantidad) as total FROM secuestros GROUP BY municipio ORDER BY total DESC LIMIT 5")
```

```python
ejecutar_sql("""
SELECT municipio, tipo_delito, SUM(cantidad) AS total
FROM secuestros
GROUP BY municipio, tipo_delito
ORDER BY municipio, total DESC
""")
```

```python
ejecutar_sql("""
SELECT 
    substr(fecha_hecho, 1, 4) AS anio,
    tipo_delito,
    SUM(cantidad) AS total
FROM secuestros
GROUP BY anio, tipo_delito
ORDER BY anio, total DESC
""")
```
